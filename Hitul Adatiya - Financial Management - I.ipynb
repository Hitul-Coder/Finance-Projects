1. 1. Expense Ratio

➢ The expense ratio is the ratio of the total fund’s expenses to its assets and reflects 
the per-unit cost of managing a fund. 
➢ Subtracted from the funds' total earnings before it is distributed to the investors. 
➢ It is an essential factor to be considered while selecting a fund since the higher 
the expense ratio, the lower is the return and vice versa. 
➢ Though SEBI has capped mutual fund's total expense ratio, still its best lower the 
better, unless we get some excellent return by paying higher expenses for fund 
management. 
➢ Expense ratios reflects the value-for-money aspect of a fund. For example, if your 
fund generates 16% returns and the TER is 2%, the NAV (Net Asset Value) of the 
fund would have grown by 14%, reflecting the returns for the investors. 
➢ Further, active funds tend to have a higher expense ratio, on account of higher 
fund management charges, which are marginal in passive funds 
and ETF (Exchange Traded Fund).


2. Fund Performance Vs. Benchmark performance (Alpha)

 It is always advisable to compare the fund performance against the benchmark.
➢ The benchmark acts as a standard for funds’ performance. If your fund is 
outperforming the benchmark consistently, it is a sign that the fund is doing 
well.
➢ Alpha represents the outperformance of the mutual fund scheme, as compared 
to the benchmark returns.
➢ The Alpha value of 0 would indicate that the fund has performed in line with the 
benchmark. 
➢ While a negative value would mean it has underperformed as compared to its 
benchmark index, a figure above 0 would mean that the fund has outperformed. 
➢ For instance, if a mutual fund generates a return of 15% in a year, whereas the 
benchmark grows at 12%, the Alpha value, in this case, would be 3. 
➢ It is generally considered as a measure that represents the value that a fund 
manager adds or subtracts to a portfolio’s returns.


3. Beta 
➢ Beta is another statistical measure calculated using regression analysis, reflecting 
the volatility of a portfolio compared to the market. 
➢ It shows the tendency of a portfolio's return to fluctuate as per the market 
movements. 
➢ Beta value of 1 indicates that the mutual fund is as volatile as its benchmark. 
➢ While a value above 1 indicates that the fund is more volatile, a value below 
represents that the fund reacts lesser than its benchmark.

4. Sharpe ratio 
➢ It Measures the mutual fund’s performance against the risk taken.
➢ Sharpe ratio is the measure of risk-adjusted return of a financial portfolio.
➢ A portfolio with a higher Sharpe ratio is considered superior relative to its peers.
➢ Sharpe ratio is a measure of excess portfolio return over the risk-free rate relative 
to its standard deviation. 
➢ If two funds offer similar returns, the one with higher standard deviation will 
have a lower Sharpe ratio. 
➢ In order to compensate for the higher standard deviation, the fund needs to 
generate a higher return to maintain a higher Sharpe ratio. 
➢ In simple terms, it shows how much additional return an investor earns by taking 
additional risk. Intuitively, it can be inferred that the Sharpe ratio of a risk-free 
asset is zero.


5. Long term returns 
➢ Mutual funds need to disclose the 1-year, 3-year, 5-year & since inception returns 
for the schemes. 
➢ While the investors may get excited with the strong 1-year returns, the 
performance review should ideally capture the long-term returns, since that may 
be a better reflection of the long-term investing decisions.

### Volatility

* Volatility has a number of different meanings.
* We use it in relation to price movement.
* Can be thought as the equivalent as risk.
* Here we will use: Average True Range (ATR)
* It is a technical analysis indicator that measures market volatility by decomposing 
  the entire range of an asset price.
* The calculation of the average true range is 14-period based. The period can be intraday, 
  daily, weekly, or monthly.
* For calculating the average true range, a series of true ranges needs to be calculated first.
* For a specific trading period, the true range is the maximum of absolute values of the following:
  - 1. Current high – Current low
  - 2. Current low – Previous close (Absolute)
  - 3. Current high – Previous close (Absolute)
* Interpretation
  - If the average true range is expanding, it implies increasing volatility in the market. 
    The average true range is non-directional; hence, an expanding range can be an indication 
    of either short sale or long buy.
  - A low value of average true range indicates small ranges in a number of consecutive periods. 
    The low average true range values imply lower price volatility.

### Sharpe Ratio

* The ratio describes how much excess return you receive for the extra volatility you endure for holding a riskier asset.
* Remember, you need compensation for the additional risk you take for not holding a risk-free asset.
* In other words, the Sharpe Ratio is the difference between the risk-free return and the return of an investment divided     by the investment’s standard deviation.
* The higher the ratio, the greater the investment return relative to the amount of risk taken, and thus, 
  the better the investment.
* The ratio can be used to evaluate a single stock or investment, or an entire portfolio.


### Sharpe Ratio Grading Thresholds:

Less than 1 :    Bad
1 – 1.99 :       Adequate/good
2 – 2.99 :       Very good
Greater than 3 : Excellent

As volatility increases, the expected return has to go up significantly to compensate for that additional risk.

### Monte Carlo Simulation and Efficient Frontier

* Applying Monte Carlo Simulation on portfolios using Sharpe Ratio (refer above Sharpe ratio calculation)
* Creating a visual Efficient Frontier based on Sharpe Ratio.
* We utilize the Monte Carlo Simulation for find the Optimum Portfolio, which is nothing but the 
  optimum mix of Risk & Return.
* Monte Carlo Simulation, also known as multiple probability simulation, is a computerized mathematical technique, which is   used to estimate the possible outcomes of an uncertain event.
* At its core, the Monte Carlo method is a way to use random samples of parameters to explore the behavior of a complex       system.
* Monte Carlo Simulation simulate random nos. and bring RANDOMNESS in model.
* Monte Carlo simulation performs risk analysis by building models of possible results by substituting a range of values—     called a probability distribution—for any factor that has inherent uncertainty. It then calculates results over and over,   each time using a different set of random values from the input probability distributions.
* Common Probability Distributions Include - Normal, Lognormal, Uniform, PERT, Discrete etc.

``There are literally infinite combinations of portfolios. To find the optimum one is a Herculean task. Monete Carlo Simulation proves to be useful as it generates the various portfolio combinations and calculates the Sharpe Ratios. Probably, this is the only way to find the Optimum Portfolio which perfectly strikes the balance between Risk & Return and maximizes the Sharpe Ratio.``

```python
# Now, we will calculate the exp. return

exp_rtns = np.sum(log_returns.mean() * weight) * 252


# Note

# 1. Firstly we calculated the daily return of portfolio using mean() function.
# 2. We multiplied it by weight (weight allocation to each stock of portfolio)
# 3. Then, we compute the annualized return by multiplying the daily returns with trading days in a year.
# 4. And lastly we added the returns of individual stocks of portfolio 
#    to arrive at a single portfolio returns.
```

```python
exp_rtns
```

```python
# Next step is to compute the Volatilty (risk)


exp_vols = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))

# Note

# 1. Refer formula above.
# 2. Dot product -  NumPy dot() function is used to return the dot product of given arrays. 
#    It accepts two arrays as arguments and calculates their dot product. 
#    It can handle 2-D arrays but considers them as matrix and will perform matrix multiplication.
```

```python
# Sharpe ratio


sharpe_ratio = exp_rtns / exp_vols

# Note

# 1. Sharpe ratio = exp. portfolio return / exp. portfolio volatility (risk - free return ignored)
```

```python
sharpe_ratio
```

```python
# Monte Carlo Simulation

n = 5000

weights = np.zeros((n, 4))
exp_rtns = np.zeros(n)
exp_vols = np.zeros(n)
sharpe_ratios = np.zeros(n)

for i in range(n):
    weight = np.random.random(4)
    weight /= weight.sum()
    weights[i] = weight
    
    exp_rtns[i] = np.sum(log_returns.mean()*weight)*252
    exp_vols[i] = np.sqrt(np.dot(weight.T, np.dot(log_returns.cov()*252, weight)))
    sharpe_ratios[i] = exp_rtns[i] / exp_vols[i]
```
